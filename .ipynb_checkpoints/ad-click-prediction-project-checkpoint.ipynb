## 1. Problem Definition

We will be exploring a **binary classification** problem. The goal is to determine the primary factor that led to ad clicks from within the dataset.

## 2. Data

The dataset came from Kaggle and is a mock dataset. It can be found here: https://www.kaggle.com/datasets/gabrielsantello/advertisement-click-on-ad

## 3. Evaluation

The goal is to gain a 85-90% accuracy rating in this proof of concept to further pursue working on the project.

## 4 . Features

The data dictionary below outlines the features within the dataset as well as the target variable "Clicked on Ad" that we want to make a prediction based on.

### Data Dictionary 

* 'Daily Time Spent on Site': consumer time on site in minutes
* 'Age': customer age in years
* 'Area Income': Avg. Income of geographical area of consumer
* 'Daily Internet Usage': Avg. minutes a day consumer is on the internet
* 'Ad Topic Line': Headline of the advertisement
* 'City': City of consumer
* 'Male': Whether or not consumer was male
* 'Country': Country of consumer
* 'Timestamp': Time at which consumer clicked on Ad or closed window
* 'Clicked on Ad': 0 or 1 indicated clicking on Ad

### Preprocessing

We'll remove the categorical columns from the dataset and proceed to experiment with model training. We'll also remove the time stamp for the time being but may use it again later. Our current look at the data has the following information:

* There is a semi-strong negative correlation between the amount of time someone spends on the internet, as well as the amount of time someone spends on the site and whether or not they clicked the ad
* There is a moderate positive correlation between people of older age and clicking the ad
* There's a moderate negative correlation between users living in areas with higher average income and whether or not they clicked the ad

```python
# Create new dataframe with categorical features removed from the data:
df_new = df.drop(['Ad Topic Line', 'City', 'Country', 'Timestamp'], axis=1)
df_new.head()
```

```python
# Set up X and y data
X = df_new.drop('Clicked on Ad', axis=1)
y = df_new['Clicked on Ad']

# Set random seed for reproducability
np.random.seed(42)

# Create train and test splits
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train
```

```python
y_train
```

```python
# Test LinearSVC model on the data
model = LinearSVC()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

### Initial model test results for LinearSVC:

The LinearSVC model trained on the data delivered an accuracy score of 0.45. This is well below our goal of 0.85 to 0.9 for this model. Let's test out the other models. We'll test the RandomForestClassifier next.

```python
# Set random seed for reproducability
np.random.seed(42)

# Test RandomForestClassifier model on the data
model = RandomForestClassifier()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

### Initial model test results for RandomForestClassifier

The RandomForestClassifier did exceptionally well when tested on the test data. 0.93 definitely meets our threshold. Let's try a cross validation score to ensure that there's no overfitting going on.

```python
# Test RandomForestClassifier using cross_val_score to reduce chance of overfitting
rf_cv_score = cross_val_score(model,
                              X,
                              y,
                              cv=5,
                              scoring="accuracy")

rf_cv_score
```

```python
# Find average of cross validated scores
rf_cv_score = np.mean(rf_cv_score)
rf_cv_score
```

### Cross Validated Accuracy of RandomForestClassifier

We have a 95% cross validated accuracy for the model. This still meets our requirements for moving forward with the model. We will now analyze the coefficient metrics to interpret the feature importance of the data

```python
# Set random seed
np.random.seed(42)

# Fit model to train data
model.fit(X_train, y_train)

# Get coefficient metrics from trained model
model.feature_importances_
```

```python
# Match features to column names
features_dict = dict(zip(df.columns, list(model.feature_importances_)))
features_dict
```

```python
features_df = pd.DataFrame(features_dict, index=[0])
features_df.T.plot.bar(title="Feature Importance", legend=False);
```

```python
# Plot ROC curve
RocCurveDisplay.from_estimator(estimator=model,
                               X=X_test,
                               y=y_test);
```
